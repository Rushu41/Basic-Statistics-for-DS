```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mlt
import seaborn as sns
```

```python
ds = pd.read_csv("titanic.csv")
```

```python
ds.head(3)
```

```python
ds["Age"].mean()
```

```python
mean = np.mean(ds["Age"])
```

```python
sns.histplot(x="Age", data = ds)
mlt.show()
```

```python
sns.histplot(x="Age", data = ds, bins=[i for i in range(0,81,10)])
mlt.show()
```

```python
sns.histplot(x="Age", data = ds, bins=[i for i in range(0,81,10)])
mlt.plot([mean for i in range(0,300)], [i for i in range(0,300)], c= "red")
mlt.show()
```

```python
ds["Age"] = ds["Age"].fillna(ds["Age"].mean())
```

```python
np.median(ds["Age"])
```

```python
np.median(ds["Fare"])
```

```python
faremedian = np.median(ds["Fare"])
```

```python
sns.histplot(x="Fare", data = ds, bins=[i for i in range(0,81,10)])
mlt.plot([faremedian for i in range(0,300)], [i for i in range(0,300)], c= "red")
mlt.plot([mean for i in range(0,300)], [i for i in range(0,300)], c= "blue")

mlt.show()
```

```python
ds["Fare"].mode()[0]
```

```python
ds["Fare"].value_counts()
```

```python
mode = ds["Fare"].mode()[0]
```

```python
sns.histplot(x="Fare", data = ds, bins=[i for i in range(0,81,10)])
mlt.plot([faremedian for i in range(0,300)], [i for i in range(0,300)], c= "red",label = "median")
mlt.plot([mean for i in range(0,300)], [i for i in range(0,300)], c= "blue",label = "mean")
mlt.plot([mode for i in range(0,300)], [i for i in range(0,300)], c= "green",label = "mode")
mlt.legend()
mlt.show()
```
