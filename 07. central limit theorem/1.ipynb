```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pop_data = []
for i in range(10000):
    n = np.random.randint(10,1000)
    pop_data.append(n)

pop_data
```

```python
pop_table =pd.DataFrame({"Pop_Data": pop_data})
pop_table
```

```python
np.mean(pop_data)
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(x="Pop_Data",data=pop_table)
plt.show()
```

```python
sample_mean = []
for no_sample in range (50):
    sample_data = []
    for data in range(500):
        sample_data.append(np.random.choice(pop_data))
    sample_mean.append(np.mean(sample_data))
    
```

```python
sample_table = pd.DataFrame({"Sample_Table": sample_mean})
```

```python
plt.figure(figsize=(3,3))
sns.kdeplot(x="Sample_Table",data=sample_table)
plt.show()
```
